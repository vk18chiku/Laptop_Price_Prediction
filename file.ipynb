```python
import re

def process_memory(x):
    sizes = re.findall(r'(\d+\.?\d*)(GB|TB)', x)
    
    total_size = 0
    for size, unit in sizes:
        size = float(size)
        if unit == 'TB':
            size *= 1024
        total_size += size
    
    return int(total_size)

# Get memory size
df['Memory_Size'] = df['Memory'].apply(process_memory)

# Create binary columns for each memory type with their numerical values
def get_ssd_value(x):
    if 'ssd' in x.lower():
        return process_memory(x)
    return 0

def get_hdd_value(x):
    if 'hdd' in x.lower():
        return process_memory(x)
    return 0

def get_hybrid_value(x):
    if 'hybrid' in x.lower():
        return process_memory(x)
    return 0

def get_flash_value(x):
    if 'flash' in x.lower():
        return process_memory(x)
    return 0

def get_other_value(x):
    x_lower = x.lower()
    if not any(t in x_lower for t in ['ssd', 'hdd', 'hybrid', 'flash']):
        return process_memory(x)
    return 0

# Apply
df['ssd'] = df['Memory'].apply(get_ssd_value)
df['hdd'] = df['Memory'].apply(get_hdd_value)
df['hybrid'] = df['Memory'].apply(get_hybrid_value)
df['flash'] = df['Memory'].apply(get_flash_value)
df['other'] = df['Memory'].apply(get_other_value)
```

```python
# Normalize OS values
def normalize_os(x):
    x_lower = x.lower()
    if 'windows' in x_lower:
        return 'Windows'
    elif 'mac' in x_lower:
        return 'Mac'
    elif 'no os' in x_lower or 'chrome' in x_lower:
        return 'Others'
    else:
        return x

df['OpSys'] = df['OpSys'].apply(normalize_os)
df['OpSys'].unique()
```

```python
clean_df=df.copy()
```

```python
x= clean_df.drop(columns='Price')
y=np.log(clean_df['Price'])
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train , y_test= train_test_split(x,y,test_size=0.10, random_state=21)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
#MODEL TRAINING

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import * 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
clean_df.info()
```

```python
# Get categorical column indices dynamically
cat_col_indices = [x.columns.get_loc(col) for col in categorical_columns]

step1=ColumnTransformer(transformers=[
    ('One_hot', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), cat_col_indices)
],remainder='passthrough')

step2= LinearRegression()

pipe= Pipeline([
    ('step1', step1),
    ('step2',step2)
])

pipe.fit(x_train, y_train)

print("✅ Pipeline model trained successfully!")
print(f"Training Score: {pipe.score(x_train, y_train):.4f}")
print(f"Testing Score: {pipe.score(x_test, y_test):.4f}")
```

```python
# DecisionTree Pipeline
step1_dt = ColumnTransformer(transformers=[
    ('One_hot', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), cat_col_indices)
],remainder='passthrough')

step2_dt = DecisionTreeRegressor(random_state=42, max_depth=15)

pipe_dt = Pipeline([
    ('step1', step1_dt),
    ('step2', step2_dt)
])

pipe_dt.fit(x_train, y_train)

print("✅ DecisionTree Pipeline model trained successfully!")
print(f"Training Score: {pipe_dt.score(x_train, y_train):.4f}")
print(f"Testing Score: {pipe_dt.score(x_test, y_test):.4f}")
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score

# Cross Validation Score
cv_scores = cross_val_score(pipe_dt, x_train, y_train, cv=5, scoring='r2')

# R² Score
y_train_pred_dt = pipe_dt.predict(x_train)
y_test_pred_dt = pipe_dt.predict(x_test)

r2_train = r2_score(y_train, y_train_pred_dt)
r2_test = r2_score(y_test, y_test_pred_dt)

print("=" * 60)
print("📊 DecisionTree Model - Cross Validation & R² Scores")
print("=" * 60)
print(f"\n🔄 Cross Validation Scores (5-Fold): {cv_scores}")
print(f"   Mean CV Score: {cv_scores.mean():.4f} (+/- {cv_scores.std():.4f})")
print(f"\n📈 R² Scores:")
print(f"   Training R²: {r2_train:.4f}")
print(f"   Testing R²:  {r2_test:.4f}")
print("=" * 60)
```

```python
# RandomForest Pipeline
step1_rf = ColumnTransformer(transformers=[
    ('One_hot', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), cat_col_indices)
],remainder='passthrough')

step2_rf = RandomForestRegressor(n_estimators=100, random_state=42, max_depth=15, n_jobs=-1)

pipe_rf = Pipeline([
    ('step1', step1_rf),
    ('step2', step2_rf)
])

pipe_rf.fit(x_train, y_train)

print("✅ RandomForest Pipeline model trained successfully!")
print(f"Training Score: {pipe_rf.score(x_train, y_train):.4f}")
print(f"Testing Score: {pipe_rf.score(x_test, y_test):.4f}")

# Cross Validation Score aur R² Score
cv_scores_rf = cross_val_score(pipe_rf, x_train, y_train, cv=5, scoring='r2')

y_train_pred_rf = pipe_rf.predict(x_train)
y_test_pred_rf = pipe_rf.predict(x_test)

r2_train_rf = r2_score(y_train, y_train_pred_rf)
r2_test_rf = r2_score(y_test, y_test_pred_rf)

print("\n" + "=" * 60)
print("📊 RandomForest Model - Cross Validation & R² Scores")
print("=" * 60)
print(f"\n🔄 Cross Validation Scores (5-Fold): {cv_scores_rf}")
print(f"   Mean CV Score: {cv_scores_rf.mean():.4f} (+/- {cv_scores_rf.std():.4f})")
print(f"\n📈 R² Scores:")
print(f"   Training R²: {r2_train_rf:.4f}")
print(f"   Testing R²:  {r2_test_rf:.4f}")
print("=" * 60)
```

```python
import xgboost as xgb

# XGBoost Pipeline
step1_xgb = ColumnTransformer(transformers=[
    ('One_hot', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), cat_col_indices)
],remainder='passthrough')

step2_xgb = xgb.XGBRegressor(n_estimators=135,
  max_depth=8,
  learning_rate=0.08889235810798346,
  subsample=0.7156405183033205,
  colsample_bytree=0.6523790762424088,
  min_child_weight=6,
  gamma=0.0051801703605352314,
  reg_alpha=0.7393878071036433,
  reg_lambda=0.7833909775555951)

pipe_xgb = Pipeline([
    ('step1', step1_xgb),
    ('step2', step2_xgb)
])

pipe_xgb.fit(x_train, y_train)

print("✅ XGBoost Pipeline model trained successfully!")
print(f"Training Score: {pipe_xgb.score(x_train, y_train):.4f}")
print(f"Testing Score: {pipe_xgb.score(x_test, y_test):.4f}")

# Cross Validation Score aur R² Score
cv_scores_xgb = cross_val_score(pipe_xgb, x_train, y_train, cv=5, scoring='r2')

y_train_pred_xgb = pipe_xgb.predict(x_train)
y_test_pred_xgb = pipe_xgb.predict(x_test)

r2_train_xgb = r2_score(y_train, y_train_pred_xgb)
r2_test_xgb = r2_score(y_test, y_test_pred_xgb)

print("\n" + "=" * 60)
print("📊 XGBoost Model - Cross Validation & R² Scores")
print("=" * 60)
print(f"\n🔄 Cross Validation Scores (5-Fold): {cv_scores_xgb}")
print(f"   Mean CV Score: {cv_scores_xgb.mean():.4f} (+/- {cv_scores_xgb.std():.4f})")
print(f"\n📈 R² Scores:")
print(f"   Training R²: {r2_train_xgb:.4f}")
print(f"   Testing R²:  {r2_test_xgb:.4f}")
print("=" * 60)
```
